```python
from histocartography.preprocessing import MacenkoStainNormalizer, VahadaneStainNormalizer, GaussianTissueMask
import glob
import numpy as np  
import matplotlib.pyplot as plt 
from PIL import Image
import os
```

# SICAPv2 dataset (WSI)

## Downsampling = 2

```python
IMAGES_PATH = 'img_sicap/'
TARGET_PATH = 'data/target_sicap.png'

image_fnames = glob.glob(os.path.join(IMAGES_PATH,'*.png'))
image_fnames.sort()

normalizer = MacenkoStainNormalizer(target_path=TARGET_PATH)
tissue_detector = GaussianTissueMask(downsampling_factor=2)
fig, ax = plt.subplots(2, 2, figsize=(10, 10))

for i, image_path in enumerate(image_fnames):
    image = np.array(Image.open(image_path))
    normalized_image = normalizer.process(image)
    ax[i, 0].imshow(normalized_image)

    tissue_mask = tissue_detector.process(normalized_image)
    ax[i, 1].imshow(normalized_image)


ax[0, 0].set_title('Normalized image before Tissue Mask generation') 
ax[0, 1].set_title('Normalized image after Tissue Mask generation') 
plt.tight_layout()
```

# Brain Dataset (Patches 256x256)

## Downsampling = 2

```python
IMAGES_PATH = 'img_patches/'
TARGET_PATH = 'data/target_patches.png'

image_fnames = glob.glob(os.path.join(IMAGES_PATH,'*.png'))
image_fnames.sort()

normalizer = MacenkoStainNormalizer(target_path=TARGET_PATH)
tissue_detector = GaussianTissueMask(downsampling_factor=2)
fig, ax = plt.subplots(2, 3, figsize=(18, 10))

for i, image_path in enumerate(image_fnames):
    image = np.array(Image.open(image_path))[:, :, :3]
    ax[i, 0].imshow(image)

    normalized_image = normalizer.process(image)
    ax[i, 1].imshow(normalized_image)

    tissue_mask = tissue_detector.process(normalized_image)
    ax[i, 2].imshow(normalized_image)

ax[0, 0].set_title('Original image before Tissue Mask generation') 
ax[0, 1].set_title('Normalized image before Tissue Mask generation') 
ax[0, 2].set_title('Normalized image after Tissue Mask generation') 
plt.tight_layout()
```
